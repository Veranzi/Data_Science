```python
import pandas as pd
import re

# Define the regular expressions for full-time and half-time results
full_time_regex = re.compile(r'Full-time Match ends,.*?(\d+),.*?(\d+)\.')
half_time_regex = re.compile(r'Half-time First Half ends,.*?(\d+),.*?(\d+)\.')

# Function to extract results
def extract_results(text):
    full_time_matches = full_time_regex.findall(text)
    half_time_matches = half_time_regex.findall(text)

    full_time_results = [f"Arsenal {match[0]} - Newcastle United {match[1]}" for match in full_time_matches]
    half_time_results = [f"Arsenal {match[0]} - Newcastle United {match[1]}" for match in half_time_matches]

    return full_time_results, half_time_results

# Sample normalized_df DataFrame (for demonstration purposes, replace this with your actual DataFrame)
normalized_df = {
    'Match': [
        'Full-time Match ends, Arsenal 0, Newcastle United 0. Second Half ends, Arsenal 0, Newcastle United 0. Half-time First Half ends, Arsenal 0, Newcastle United 0.',
        'Full-time Match ends, Chelsea 2, Liverpool 1. Second Half ends, Chelsea 2, Liverpool 1. Half-time First Half ends, Chelsea 1, Liverpool 0.',
        # Add more rows as needed
    ]
}
normalized_df = pd.DataFrame(normalized_df)

# Apply function to the 'Match' column
normalized_df[['Full-Time Results', 'Half-Time Results']] = normalized_df['Match'].apply(lambda x: pd.Series(extract_results(x)))

# Display the updated DataFrame
print(normalized_df)
```

```python
import pandas as pd
# Load the players information dataset
df_players = pd.read_csv('player_info.csv')
#Loading the match dataset
# Read the JSON file into a DataFrame
json_file_path = '/content/epl_2022_2023_07_02_2023.json'
df_match_data = pd.read_json(json_file_path)
```

```python
print(df_players.head(5))
print(df_match_data.head(5))
```

```python
#Transposing the Match Dataset
# Transpose rows and columns
df = df_match_data.transpose()

df
```

```python
#Resetting the index
df.reset_index()
```

```python
# Rename columns
df.rename(columns={'index': 'ID',
                   'team1_name': 'home',
                   'team1_startings':'home_start',
                   'team1_subs':'home_subs',
                   'team1_stat':'home_stat',
                   'team2_name':'away',
                   'team2_startings':'away_start',
                   'team2_subs':'away_subs',
                   'team2_stat': 'away_stat'}, inplace=True)

df
```

```python

# Split the 'away' column into 'away_possession' and 'away_shots_on_target'
df['home_possession'] = df['home_stat'].apply(lambda x: x['possession_%'])
df['home_shots_on_target'] = df['home_stat'].apply(lambda x: x['shots_on_target'])

# Drop the original 'home_stat' column
df.drop(columns=['home_stat'], inplace=True)
df
```

```python
# Split the 'away' column into 'away_possession' and 'away_shots_on_target'
df['away_possession'] = df['away_stat'].apply(lambda x: x['possession_%'])
df['away_shots_on_target'] = df['away_stat'].apply(lambda x: x['shots_on_target'])

# Drop the original 'away_stat' column
df.drop(columns=['away_stat'], inplace=True)


df
```

```python
# Drop the 'matchweek' column
df.drop(columns=['matchweek'], inplace=True)

print(df)
```

```python
# Set display options to show full text
pd.set_option('display.max_colwidth', None)

# Display the DataFrame with full text in the 'event' column
print(df[['event']])
```

```python
data=df

# Function to determine match outcome
def determine_outcome(event):
    # Remove "Full-time Match ends" and split the event string into components
    parts = event.replace('Full-time Match ends, ', '').split(',')

    # Extract home and away team names and scores
    home_team_score = parts[0].split()
    home_team = ' '.join(home_team_score[:-1])  # Extract all elements except the last one as the team name
    home_score = ''.join(filter(str.isdigit, home_team_score[-1]))  # Extract only digits from the score
    home_score = int(home_score) if home_score else 0  # Convert score to integer, defaulting to 0 if no digits found

    away_team_score = parts[1].split()
    away_team = ' '.join(away_team_score[:-1])  # Extract all elements except the last one as the team name
    away_score = ''.join(filter(str.isdigit, away_team_score[-1]))  # Extract only digits from the score
    away_score = int(away_score) if away_score else 0  # Convert score to integer, defaulting to 0 if no digits found

    # Determine match outcome
    if home_score > away_score:
        outcome = 'Home Win'
    elif home_score < away_score:
        outcome = 'Away Win'
    else:
        outcome = 'Draw'

    return outcome

# Apply the function to the event column
data['outcome'] = data['event'].apply(determine_outcome)

# Display the DataFrame with outcomes
print(data)
data.tail()
```
